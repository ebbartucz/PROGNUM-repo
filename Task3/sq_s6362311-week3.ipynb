Question 3.2:

```python
from scipy.constants import G 
from math import sqrt

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

velocities = []
for M, r in zip(masses, radii):
    vesc = sqrt(2*G*M/r**2)
    velocities.append(vesc)
    print(f"The mass of the object: {M:.2e}")
    
print(velocities)
print()

dictionary = {}
for i in range(len(names)):
    dictionary[names[i]] = velocities[i] # calling the dictionary by the keys
print(dictionary)
```

Question 3.10:

```python
def cylinder(r,h):
    from math import pi
    area = 2*pi*r**2+2*pi*r*h
    volume = pi*r**2*h
    return area, volume


area,volume=cylinder(3,4)
print(f"The area and volume of the cylinder:{area,volume}")
print()
ratio = area/volume
print(f"The ratio is: {ratio}")
```

```python
Sum = [] # A list that will contain partial sums of the series
N=[100,200,300,400,500] # The indeces until which partial sums are calculated
for n in range(len(N)):
    Sum.append(sum(1/(x**2) for x in range(1,N[n]+1))) # The method of calculating the partial sums

if (Sum[-1]-Sum[-2]) < (Sum[-2]-Sum[-3]): # If the difference between partial sums decreases as N becomes larger
    print("The series converges.")
else:
    print("The series does not converge.")
print(Sum)
```

Question 3.17:

```python
import numpy as np
from math import sqrt

def stdev(y):
    N = len(y)
    mu = np.array(1/N*np.sum(y))
    sig = sqrt(1/N*np.sum((y-mu)**2))
    return sig

y = np.random.normal(0,2,10000)

s1 = stdev(y)

s2 = np.std(y)
print(s1,s2)
print()

#One-liner:

def stdev1(y):
    sig = sqrt(1/len(y)*np.sum((y-np.array(1/len(y)*np.sum(y)))**2))
    
    return sig
s = stdev1(y)
print(s)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *

x = list(range(-5,6,1))
y = []
for i in range(len(x)):
    y.append(cosh(x[i]))
plt.scatter(x, y, label='catenary')
plt.title("Catenary plot using range")
plt.grid()
plt.xlabel("x")
plt.ylabel("cosh(x)")
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11)
plt.show()

x = np.arange(-5,6,1)
y = np.cosh(x)
plt.scatter(x, y, label='catenary')
plt.title("Catenary plot using arange")
plt.grid()
plt.xlabel("x")
plt.ylabel("cosh(x)")
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11)
plt.show()

```

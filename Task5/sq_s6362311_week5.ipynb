Question 5.2 : Functions loadtxt() and savetxt()

```python
import numpy as np

filename = input("Please enter a valid file name ending in .txt:")
file = np.loadtxt(filename)

x = file[:,0]
y = file[:,1]

print(f"x={x}")
print(f"y={y}")
print()

print(f"The maximum value of x:{x.max()}")
print(f"The maximum value of y:{y.max()}")
print(f"The minimum value of x:{x.min()}")
print(f"The minimum value of y:{y.min()}")
print(f"The average value of x:{x.mean()}")
print(f"The average value of y:{y.mean()}")
print()

ysort = np.sort(y) # sorting the array to find the median
y_med = (ysort[len(ysort)//2]+ysort[len(ysort)-(len(ysort)//2)])/2 # calculating the average of the two central elements
print()

y2 = (y - y_med)**2

print(f"The new array: {y2}")
```

Question 5.6 : Array concatenation

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(f"The concatenation of the two lists x and y: {np.concatenate((a, b))}")
# it is a one dimensional list/array

print()


a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print(f"np.concatenate((a, b)): {np.concatenate((a, b))}") # axis = 0 is default
print(f"np.concatenate((a, b), axis=0): {np.concatenate((a, b), axis=0)}") # axis = 0 concatenates vertically
print(f"np.concatenate((a, b), axis=1): {np.concatenate((a, b), axis=None)}") # axis = 1 concatenates horizontally
```

Question 5.10 : Slicing in a multi dimensional array

```python
import numpy as np
import matplotlib.pyplot as plt

file = np.loadtxt('ngc6946.txt')
file = file.reshape(40,50,70)

filematrix = file[:,25,35]
print(filematrix)
print()

print(f"Max: {filematrix.max()}")
print(f"Min: {filematrix.min()}")
print(f"Sum: {filematrix.sum()}")
print(f"Mean: {filematrix.mean()}")
print(f"Standard deviation: {filematrix.std()}")

plt.plot(range(len(filematrix)),filematrix)
plt.rc('text', usetex = True)
plt.title("z profile of the data of NGC 6946 at (25,35)")
plt.text(20,0.04, s = f"The maximum: {filematrix.max():.3f} \n The minimum: {filematrix.min():.3f} \n The mean: {filematrix.mean():.3f} \n The sum: {filematrix.sum():.3f} \n The standard deviation: {filematrix.std():.3f}")
plt.show()
```

Question 5.12 : Getting acquainted with classes

```python
class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    from scipy.constants import c
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        v = self.redshift()*self.c/(1000)
        return v

gal1 = Galaxy("NGC1357",4000)

print(f"{gal1.velocity()} km/s")
```

Question 5.13 : Ellipse plotter with astronomical parameters

```python
from matplotlib.patches import Ellipse

class AstroEllipse(Ellipse):
   def __init__(self, xc, yc, semi_major, semi_minor, position_angle, **args):
      # We use the (sort of) astronomical angle where 'angle' defines)
      # the position angle of the major axis of a ring projected onto the sky
      # (i.e. an ellipse). This is an angle taken in anti-clockwise direction between
      # the north direction on the sky and the major axis of the receding half of the
      # galaxy (Rots 1975, astron, astrophys 45, 43).
      # Here we assume (!) north is in the positive Y direction.
      # If not, you need to correct the angle for the rotation of the image.
      # Parameter 'position_angle' enters in degrees
      assert(semi_major >= semi_minor)  # Prerequisite to make angle unambiguous
      XY = (xc, yc)                  # MPL requires center as a tuple
      angle_mpl = position_angle     # Angle for MatPlotLib's ellipse patch
      height = 2.0*semi_major        # height: is along vertical axis for an unrotated ellipse
      width  = 2.0*semi_minor        # width:  is along horizontal axis for an unrotated ellipse
      super().__init__(XY, width, height, angle=angle_mpl, **args)
    
semi_major = 100
semi_minor = 50
position_angle = 30 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, show
Nx = 400
Ny = 300
data = np.random.random((Ny,Nx))


fig = figure()
ellipse = AstroEllipse((Nx/2),(Ny/2),semi_major,semi_minor,position_angle,fill=False, color = "purple",linestyle = '--', linewidth = 5)
print(ellipse)
frame = fig.add_subplot(1,1,1, aspect=1)
frame.add_patch(ellipse)
plt.imshow(data, interpolation='none', origin="lower", cmap='gray')
plt.show()

```

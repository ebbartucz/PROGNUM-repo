```python
from math import atan2, degrees
x=float(input(f"The x coordinate is:"))
y=float(input(f"The y coordinate is:"))

theta = degrees(atan2(y,x))
print("The angle from the x axis is:",theta, "°")

if x >= 0 and y >= 0:
    q = 1
elif x < 0 and y > 0:
    q = 2
elif x < 0 and y < 0:
    q = 3
else:
    q = 4
print ("The number of the quadrant the selected point is in:", q)
```

Question 2.15:

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

n1 = set(night1)
n2 = set(night2)

print("The two sets of observations:",n1,n2)
inters = set.intersection(n1,n2)

print ("The intersection of the two sets:",inters)
print()

if 5 in inters:
    print("5 is in the intersection of the two sets.")
else:
    print("5 is NOT in the intersection of the two sets.")
    
if 5 in n1:
    print("5 is in the set of observation for the first night.")
else:
    print("5 is NOT in the set of observation for the first night.")
```
